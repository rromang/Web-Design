```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = r'Resources/cities.csv'
```

```python
weather_df = pd.read_csv(path)
```

```python
import datetime
weather_df['Date']=pd.to_datetime(weather_df['Date'],unit='s')
weather_df['Date'] = weather_df['Date'].dt.strftime('%d-%m-%Y %I:%M:%S')
```

```python
weather_df.head(15)
```

```python
sns.set_theme('notebook')

g = sns.relplot(x="Lat", y="Max Temp", hue="Max Temp", data=weather_df)
g.fig.suptitle("vs Max Temperature", y=1.10, size=20)
plt.xlabel("Latitude", size=16)
plt.ylabel("Max Temperature (ºF)", size=16)
plt.title("City Latitude vs Max Temperature (06/13/2018)", size=16)
```

```python
sns.set_theme('notebook')

g = sns.relplot(x="Lat", y="Humidity", hue="Humidity", data=weather_df)
g.fig.suptitle("vs HUmidity", y=1.10, size=20)
plt.xlabel("Latitude", size=16)
plt.ylabel("Humidity (%)", size=16)
plt.title("City Latitude vs Humidity (06/13/2018)", size=16)
```

```python
sns.set_theme('notebook')

g = sns.relplot(x="Lat", y="Wind Speed", hue="Wind Speed", data=weather_df)
g.fig.suptitle("vs Wind Speed", y=1.10, size=20)
plt.xlabel("Latitude", size=16)
plt.ylabel("Wind Speed (mph)", size=16)
plt.title("City Latitude vs Wind Speed (06/13/2018)", size=16)
```

```python
sns.set_theme('notebook')

g = sns.relplot(x="Lat", y="Cloudiness", hue="Cloudiness", data=weather_df)
g.fig.suptitle("vs Cloudiness", y=1.10, size=20)
plt.xlabel("Latitude", size=16)
plt.ylabel("Cloudiness (%)", size=16)
plt.title("City Latitude vs Cloudiness (06/13/2018)", size=16)
```
